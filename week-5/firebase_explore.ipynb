```python
import random
import datetime
from firebase_admin import db


ref = db.reference("/")
ref.child("sensors").child("DHT22_001").set({
    "sensor_type": "DHT22",
    "location": "Room_A",
    "description": "Temperature and Humidity Sensor"
})


dht22_ref = db.reference("/sensors/DHT22_001/readings")

for i in range(10):

    temperature = round(random.uniform(18.0, 35.0), 1)  
    humidity = round(random.uniform(30.0, 80.0), 1)     
    timestamp = datetime.datetime.now().isoformat()
    
    reading = {
        "timestamp": timestamp,
        "temperature": temperature,
        "humidity": humidity,
        "sensor_id": "DHT22_001"
    }
    
    dht22_ref.push().set(reading)
    print(f"Inserted: {reading}")


print("\n=== All DHT22 Readings ===")
all_readings = dht22_ref.get()
if all_readings:
    for key, value in all_readings.items():
        print(f"Key: {key}, Data: {value}")
else:
    print("No data found")
```

```python
# Create SR04 sensor structure
ref = db.reference("/sensors/SR04_001")
ref.set({
    "sensor_type": "SR04_Ultrasonic",
    "location": "Door_A", 
    "description": "Distance Measurement Sensor",
    "unit": "cm"
})

# Generate random SR04 data
sr04_ref = db.reference("/sensors/SR04_001/readings")

for i in range(10):
    # Generate realistic distance data (2cm to 400cm range for SR04)
    distance = round(random.uniform(2.0, 400.0), 1)
    timestamp = datetime.datetime.now().isoformat()
    
    reading = {
        "timestamp": timestamp,
        "distance": distance,
        "sensor_id": "SR04_001",
        "unit": "cm"
    }
    
    sr04_ref.push().set(reading)
    print(f"Inserted: {reading}")

# Query all SR04 data
print("\n=== All SR04 Readings ===")
all_readings = sr04_ref.get()
if all_readings:
    for key, value in all_readings.items():
        print(f"Key: {key}, Data: {value}")
else:
    print("No data found")

# Query both sensors
print("\n=== All Sensors Overview ===")
sensors_ref = db.reference("/sensors")
all_sensors = sensors_ref.get()
for sensor_id, sensor_data in all_sensors.items():
    print(f"Sensor: {sensor_id}")
    print(f"Type: {sensor_data.get('sensor_type')}")
    print(f"Location: {sensor_data.get('location')}")
    if 'readings' in sensor_data:
        print(f"Total readings: {len(sensor_data['readings'])}")
    print("---")
```

```python
import serial
import time
import json
import firebase_admin
from firebase_admin import credentials, db

databaseURL = 'https://iot-db-27d22-default-rtdb.asia-southeast1.firebasedatabase.app/'
cred_obj = firebase_admin.credentials.Certificate(
    'iot-db-27d22-firebase-adminsdk-fbsvc-43adb56347.json'
)
default_app = firebase_admin.initialize_app(cred_obj, {
	'databaseURL':databaseURL
	})

ref = db.reference('/gyroscope')

ser = serial.Serial('/dev/cu.usbmodem11101', 9600, timeout=1)
time.sleep(2) 

try:
    while True:
        line = ser.readline().decode().strip()
        # print("Received:", line)
        parts = line.split(',')
        if len(parts) != 3:
            continue
        x = float(parts[0].split('=')[1])
        y = float(parts[1].split('=')[1])
        z = float(parts[2].split('=')[1])
        timestamp = time.time()
        data = {
            'timestamp': timestamp,
            'gyro_x': x,
            'gyro_y': y,
            'gyro_z': z
        }
        ref.push(data)
        print("Pushed:", data)
except KeyboardInterrupt:
    ser.close()
    print("Stopped data collection.")
```
